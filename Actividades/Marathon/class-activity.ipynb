## Juntando todo

```python
import csv
from dataclasses import dataclass

@dataclass
class Runner:
    name: str = ''
    age: int = 0
    city: str = ''
    country: str = ''
    time: str = ''

file = open('marathon_results_2017.csv')
data = csv.reader(file)

categories = next(data)

runners = []
for p in data:
    runner = Runner(p[2], int(p[3]), p[5], p[7], p[-4])
    runners.append(runner)
    
runners
```

## Filtrado de los datos

```python
# Personas mayores a 35 años
mayores35 = [ runner for runner in runners if runner.age > 35 ]

# Número de corredores
print(f'En total hubo {len(runners)} competidores, {len(mayores35)} mayores a 35 años')

# Competidores mexicanos
mexicans = [ runner for runner in runners if runner.country == 'MEX' ]
print(f'Hubo {len(mexicans)} corredores mexicanos')
```

```python
import re
# Corredores tapatíos
tapatios = [ runner for runner in runners if re.search("Guadalajara|Zapopan|Tlaquepaque", runner.city) ]
tapatios
```
