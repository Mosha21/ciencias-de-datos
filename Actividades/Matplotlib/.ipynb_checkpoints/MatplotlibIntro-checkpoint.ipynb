```python
x = list(range(-10, 10))
y = [i * i for i in x]
print(y)
plt.plot(y) # No usará los valores de x
```

```python
plt.plot(x, y) # Usará los valores de x
```

```python
# Sirve para trabajar con datos numéricos
import numpy as np

x1 = list(range(-10, 10)) # python standar

x2 = np.arange(-10, 10, .5) # por escalón
x3 = np.linspace(-10, 10, 8) # cuántos números tiene el arreglo

y = x3**2
plt.plot(x3, y)
plt.plot(x3, y + 20)
plt.plot(x3, 2 * y)
```

```python

dias = np.arange(1, 8)
moscas = np.array([10, 3, 2, 4, 10, 1, 3])
plt.plot(dias, moscas)
```

## Agregar información a la gráfica

```python
plt.plot(dias, moscas)
plt.grid(True) # cuadrícula
```
