```python
# Sirve para trabajar con datos num√©ricos
import numpy as np

x1 = list(range(-10, 10)) # python standar

x2 = np.arange(-10, 10, .5) # por escal√≥n
x3 = np.linspace(-10, 10, 8) # cu√°ntos n√∫meros tiene el arreglo

y = x3**2
plt.plot(x3, y)
plt.plot(x3, y + 20)
plt.plot(x3, 2 * y)
```

```python

dias = np.arange(1, 8)
moscas = np.array([10, 3, 2, 4, 10, 1, 3])
plt.plot(dias, moscas)
```

## Agregar informaci√≥n a la gr√°fica

```python
moscas+=2

plt.plot(dias, moscas)
plt.grid(True) # cuadr√≠cula
plt.xlabel('Dias')
plt.ylabel('Moscas')
```

```python
# Elecci√≥n manual de los l√≠mites de los ejes

dias = np.arange(1, 8)
moscas = np.array([10, 3, 2, 4, 10, 1, 3])
moscas+=2
plt.plot(dias, moscas)
plt.grid(True) # cuadr√≠cula
plt.xlabel('Dias')
plt.ylabel('Moscas')
plt.title('Alimentaci√≥n del Ajolote')

xmin = 1
xmax = 7
ymin = 0
ymax = 15
plt.axis([xmin, xmax, ymin, ymax])
```

```python
# Elecci√≥n manual de los l√≠mites de los ejes

dias = np.arange(1, 8)
moscas_pancho = np.array([10, 3, 2, 4, 10, 1, 3])
moscas_juana = np.array([5, 5, 5, 6, 4, 6, 5])
plt.plot(dias, moscas_juana)
plt.plot(dias, moscas_pancho)
plt.grid(True) # cuadr√≠cula
plt.xlabel('Dias')
plt.ylabel('Moscas')
plt.title('Alimentaci√≥n del Ajolote')

plt.axis(ymin=0)
```

```python
# Marcar las diferentes curvas
# 1. agregar una etiqueta a cada gr√°fica (label)
# 2. Llamar a plt.legends()

dias = np.arange(1, 8)
moscas_pancho = np.array([10, 3, 2, 4, 10, 1, 3])
moscas_juana = np.array([5, 5, 5, 6, 4, 6, 5])
plt.plot(dias, moscas_juana, label='Juana')
plt.plot(dias, moscas_pancho, label='Pancho')
plt.grid(True) # cuadr√≠cula
plt.xlabel('Dias')
plt.ylabel('Moscas')
plt.title('Alimentaci√≥n del Ajolote')
plt.legend()

plt.axis(ymin=0)
```
